### Mapas de calor(Heatmaps)

```python
vuelos = sns.load_dataset('flights')
vuelos.head()
```

```python
vuelos_matrix = vuelos.pivot_table(index='month', columns='year', values='passengers')
vuelos_matrix
```

```python
sns.heatmap(vuelos_matrix)
```

```python
sns.heatmap(vuelos_matrix, cmap='viridis')
```

```python
sns.heatmap(vuelos_matrix, cmap='coolwarm')
```

```python
sns.heatmap(vuelos_matrix, cmap='coolwarm', linecolor='white', linewidths=2)
```

### Gráficos de cuadrícula (grids): PairGrid y FaceGrid

```python
flores = sns.load_dataset('iris')
flores.head()
```

```python
sns.pairplot(flores)
```

```python
import matplotlib.pyplot as plt
```

```python
graficos = sns.PairGrid(flores)
graficos.map(plt.scatter)
```

```python
graficos = sns.PairGrid(flores)
graficos.map_diag(sns.histplot)
graficos.map_upper(plt.scatter)
graficos.map_lower(sns.kdeplot)
```

### Estilos y Colores

```python
propinas = sns.load_dataset('tips')
propinas.head()
```

```python
sns.set_context('notebook', font_scale=1.5)
plt.figure(figsize=(10,4))
sns.set_style('white')
sns.countplot(x='sex', data=propinas)
sns.despine(left=True, bottom=True)
```

```python
sns.countplot(x='sex', data=propinas, palette='Set2')
```

```python
colores = ['lightgreen', 'blue']
sns.countplot(x='sex', data=propinas, palette=colores, hue='sex')
plt.show()
```
